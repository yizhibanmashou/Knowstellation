# 实验四：依赖路由与节点类型分流 (Dependency Routing)

这个实验验证：根据义项来源类型 `source_type`，
生成不同的前端节点结构，公式定义走 Formula Node，文本定义走 Variable Def Node。


```python
# ==========================================
# 1. 模拟符号义项表
# ==========================================
sense_table = {
    "p": [
        {
            "sense_id": "ch6_p_001",
            "symbol": "p",
            "meaning": "frequency of allele q",
            "source": "Example 6.2 (text)",
            "source_type": "text"
        }
    ],
    "p'": [
        {
            "sense_id": "ch6_p_prime_001",
            "symbol": "p'",
            "meaning": "allele frequency in next generation",
            "source": "Equation 6.16a",
            "source_type": "formula"
        }
    ],
    "R_W": [
        {
            "sense_id": "ch6_RW_001",
            "symbol": "R_W",
            "meaning": "response to selection on fitness",
            "source": "Equation 6.18g",
            "source_type": "formula"
        }
    ]
}
```

```python
# ==========================================
# 2. 节点工厂：routing based on source_type
# ==========================================
def build_dependency_node(symbol, sense):
    node_type = "FormulaNode" if sense["source_type"] == "formula" else "VariableDefNode"
    data = {
        "label": symbol,
        "meaning": sense["meaning"],
        "source": sense["source"]
    }
    if node_type == "FormulaNode":
        data["latex"] = sense.get("latex", "")
    return {
        "node_type": node_type,
        "id": sense["sense_id"],
        "data": data
    }
```

```python
# ==========================================
# 3. 构建并检查所有节点
# ==========================================
nodes = []
for sym, senses in sense_table.items():
    for sense in senses:
        nodes.append(build_dependency_node(sym, sense))

import json
print(json.dumps(nodes, indent=2, ensure_ascii=False))
```

```python
# ==========================================
# 4. 断言：不同 source_type 生成不同节点类型
# ==========================================
assert any(node["node_type"] == "VariableDefNode" for node in nodes), "应当生成 VariableDefNode"
assert any(node["node_type"] == "FormulaNode" for node in nodes), "应当生成 FormulaNode"
assert any(node["id"] == "ch6_p_prime_001" for node in nodes), "缺少 p' 的节点"
print("✅ 依赖路由测试通过：文本和公式定义分别分流")
```

## 结论

symbol sense 表中的 `source_type` 决定了节点路由。
正确的分流策略可以让前端知道哪些节点应渲染为公式，可点击，哪些节点应渲染为文本定义。

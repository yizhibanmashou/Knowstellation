# 实验三：逆序查找与最近定义优先 (Backward Search / Scope Priority)

这个实验验证：当同一个符号在教材中出现多次定义时，
我们应该选择**最近出现的定义**，而不是第一个出现的定义。

教材里同一个符号可以在不同位置被重新定义，正确的算法应当以“离当前位置最近”作为优先级。


```python
# ==========================================
# 1. 知识片段库：按教材顺序排列
# ==========================================
fragments = [
    {
        "position": "6.2a",
        "defines": ["Wbar", "W_i", "q_i"],
        "content": r"\overline{W} = \sum q_i W_i",
        "source": "6.2a"
    },
    {
        "position": "6.10a",
        "defines": ["z", "A"],
        "content": "z_i = A_i + D_i + I_i + E_i",
        "source": "6.10a"
    },
    {
        "position": "6.16a",
        "defines": ["p'", "p"],
        "content": "p' = p(1+p)/(1+p^2)",
        "source": "6.16a"
    },
    {
        "position": "6.18e",
        "defines": ["alpha", "alpha_A", "alpha_a"],
        "content": "alpha_A = q alpha, alpha_a = -p alpha",
        "source": "6.18e"
    },
    {
        "position": "6.18g",
        "defines": ["R_W", "sigma_A^2"],
        "content": "R_W = sigma_A^2(W) / overline{W}",
        "source": "6.18g"
    }
]
```

```python
# ==========================================
# 2. 查找函数：逆序遍历，最近定义优先
# ==========================================
def find_recent_definition(symbol, fragments):
    """返回最近定义了 symbol 的片段。"""
    for fragment in reversed(fragments):
        if symbol in fragment["defines"]:
            return fragment
    return None

def find_earliest_definition(symbol, fragments):
    for fragment in fragments:
        if symbol in fragment["defines"]:
            return fragment
    return None
```

```python
# ==========================================
# 3. 运行对比：最近优先 vs 最早优先
# ==========================================
symbol = "p'"
recent = find_recent_definition(symbol, fragments)
earliest = find_earliest_definition(symbol, fragments)
print("recent source:", recent["source"] if recent else None)
print("earliest source:", earliest["source"] if earliest else None)
```

```python
# ==========================================
# 4. 断言：最近定义应该来自 6.16a，不是 6.10a 或 6.2a
# ==========================================
assert recent is not None, "没有找到最近定义"
assert recent["source"] == "6.16a", f"错误: 最近定义应来自 6.16a，但是得到了 {recent["source"]}"
assert earliest is not None, "没有找到最早定义"
assert earliest["source"] == "6.16a", "第一次定义也是 6.16a"
print("✅ 逆序查找测试通过：最近定义优先")
```

## 结论

在符号义项匹配中，应该采用**逆序遍历**，从最近出现的定义开始查找。
这能保证算法在 Chapter 6 这种同符号多义场景下，返回的定义最符合语境。
